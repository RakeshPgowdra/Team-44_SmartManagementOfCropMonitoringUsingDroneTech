```python

# Load your trained models
leaf_model = load_model("leaf_classifier_cnn_fixed.h5")
# disease_model = load_model("YashModel/yashwanth3(200e)_plantmodel.keras", compile=False, safe_mode=False)

# Load model with dummy policy
disease_model = load_model("sambram_disease(150e)240_plantmodel.h5", compile=False)
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing import image
from PIL import Image
import requests
import io

# --- Load Excel Sheet ---
excel_url = "Updated_Plant_Diseases_Info_with_New_Data.xlsx"  # üîÅ Replace this with your actual Excel URL or local file path
df = pd.read_excel(excel_url)  # Ensure the Excel sheet has a 'Disease' column

# --- Disease Classes (must match model output order) ---
disease_classes = [
    "Chili_healthy",
    "Corn_commonrust"
    "Corn_grayleafspot",
    "Cotton_Aphids",
    "Cotton_Armyworm",
    "Cotton_Healthy",
    "Tomato__Bacterial_spot",
    "Tomato__Early_blight",
    "Tomato__healthy",
]

# --- Prediction Function ---
def predict_image(img_path):
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Stage 1: Leaf Detection
    is_leaf_prob = leaf_model.predict(img_array)[0][0]
    is_leaf = is_leaf_prob < 0.5

    if is_leaf:
        print(f"üü¢ Stage 1: Detected as a Leaf ({is_leaf_prob:.2f})")

        # Stage 2: Disease Prediction
        disease_probs = disease_model.predict(img_array)[0]
        predicted_index = np.argmax(disease_probs)
        predicted_disease = disease_classes[predicted_index]
        confidence = disease_probs[predicted_index]

        print(f"üåø Stage 2: Predicted Disease ‚Üí {predicted_disease} ({confidence:.2f})")

        # Match disease info in Excel
        match = df[df['Disease'] == predicted_disease]
        if not match.empty:
            print("üìã Disease Info from Excel (Key-Value format):")
            for index, row in match.iterrows():
                print("\n--- Disease Information ---")
                for col in match.columns:
                    print(f"{col}: {row[col]}")
        else:
            print("‚ö†Ô∏è No matching disease info found in the Excel sheet.")
    else:
        print(f"üî¥ Stage 1: Not a Leaf ({is_leaf_prob:.2f}) ‚Äî No further classification.")

# --- Hardcoded image path (üñºÔ∏è replace with your actual image path) ---
image_path = "LeafOrNotAll/leafNotleafDataset/test/Not_leaf/4.jpeg"  # üîÅ Replace with your test image
predict_image(image_path)
```

```python
###automatic file picker
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing import image
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from PIL import Image
import requests
import io
import tempfile

# --- Load Excel Sheet ---
# Replace with your actual Excel URL or local file path
excel_url = "Updated_Plant_Diseases_Info_with_New_Data.xlsx"  # <-- Replace this
df = pd.read_excel(excel_url)  # Make sure it has a 'Disease' column

# --- Disease Classes (Must match your model output order) ---
disease_classes = [
    "Chili_healthy",
    "Corn_commonrust",
    "Corn_grayleafspot",
    "Cotton_Aphids",
    "Cotton_Armyworm",
    "Cotton_Healthy",
    "Tomato__Bacterial_spot",
    "Tomato__Early_blight",
    "Tomato__healthy",
]

# --- Prediction Function ---
def predict_image(img_path):
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Stage 1: Leaf Detection
    is_leaf_prob = leaf_model.predict(img_array)[0][0]
    is_leaf = is_leaf_prob < 0.5

    if is_leaf:
        print(f"üü¢ Stage 1: Detected as a Leaf ({is_leaf_prob:.2f})")

        # Stage 2: Disease Prediction
        disease_probs = disease_model.predict(img_array)[0]
        predicted_index = np.argmax(disease_probs)
        predicted_disease = disease_classes[predicted_index]
        confidence = disease_probs[predicted_index]

        print(f"üåø Stage 2: Predicted Disease ‚Üí {predicted_disease} ({confidence:.2f})")

        # Match disease info in Excel
        match = df[df['Disease'] == predicted_disease]
        if not match.empty:
            print("üìã Disease Info from Excel:")
            print(match.to_string(index=False))
        else:
            print("‚ö†Ô∏è No matching disease info found in the Excel sheet.")
    else:
        print(f"üî¥ Stage 1: Not a Leaf ({is_leaf_prob:.2f}) ‚Äî No further classification.")

# --- Use File Picker ---
def choose_and_predict():
    Tk().withdraw()  # Hide root window
    img_path = askopenfilename(filetypes=[("Image files", "*.jpg *.jpeg *.png")])
    if img_path:
        predict_image(img_path)
    else:
        print("‚ùå No image selected.")

# --- (Optional) Predict from image URL ---
def predict_image_from_url(image_url):
    response = requests.get(image_url)
    img = Image.open(io.BytesIO(response.content))
    with tempfile.NamedTemporaryFile(suffix=".jpg", delete=False) as tmp:
        img.save(tmp.name)
        predict_image(tmp.name)

# üëá Run this to open the file picker and predict
choose_and_predict()

# (Optional) Run this to test from image URL
# predict_image_from_url("https://example.com/your_image.jpg")
```
